```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
INPUT_DATASET = "../dataset"
OUTPUT_BASE = "../data_split"

TRAIN_PATH = os.path.join(OUTPUT_BASE, "train")
TEST_PATH  = os.path.join(OUTPUT_BASE, "test")
```

```python

TEST_SIZE = 0.2
RANDOM_STATE = 42
```

```python
os.makedirs(TRAIN_PATH, exist_ok=True)
os.makedirs(TEST_PATH, exist_ok=True)
```

```python
for cls in os.listdir(INPUT_DATASET):
    cls_path = os.path.join(INPUT_DATASET, cls)
    if not os.path.isdir(cls_path):
        continue

    images = [
        img for img in os.listdir(cls_path)
        if img.lower().endswith((".jpg", ".png", ".jpeg"))
    ]

    train_imgs, test_imgs = train_test_split(
        images,
        test_size=TEST_SIZE,
        random_state=RANDOM_STATE
    )

    for split, imgs in zip([TRAIN_PATH, TEST_PATH], [train_imgs, test_imgs]):
        out_cls = os.path.join(split, cls)
        os.makedirs(out_cls, exist_ok=True)
        for img in imgs:
            shutil.copy(
                os.path.join(cls_path, img),
                os.path.join(out_cls, img)
            )

print("✅ Dataset split completed")
```
