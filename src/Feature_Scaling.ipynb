```python
import os
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib
```

## Paths

```python

FEATURES_PATH = r"C:\Users\pc\PycharmProjects\Project\features"
MODELS_PATH = r"C:\Users\pc\PycharmProjects\Project\models"

os.makedirs(MODELS_PATH, exist_ok=True)
```

## Load features

```python

X = np.load(os.path.join(FEATURES_PATH, "X_hog.npy"))
y = np.load(os.path.join(FEATURES_PATH, "y.npy"))

print("Loaded X shape:", X.shape)
print("Loaded y shape:", y.shape)
```

# Feature Scaling


```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Save scaled features and scaler

```python
np.save(os.path.join(FEATURES_PATH, "X_hog_scaled.npy"), X_scaled)
joblib.dump(scaler, os.path.join(MODELS_PATH, "hog_scaler.pkl"))

print("Feature scaling completed ✔")
print("Scaled X shape:", X_scaled.shape)
```
