```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
import joblib
```

## Load features

```python
X_train = np.load("../features/X_train_cnn.npy")
X_test = np.load("../features/X_test_cnn.npy")

print("Original X shape:", X_train.shape)
print("Original y shape:", X_test.shape)
```

# Feature Scaling



```python
scaler = Normalizer(norm="l2")

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
np.save("../features/X_train_scaled.npy", X_train_scaled)
np.save("../features/X_test_scaled.npy", X_test_scaled)

joblib.dump(scaler, "../models/scaler.pkl")

print(" Feature scaling completed")
```
